# 6) Tokenization

```python
# //1st process of text pre-processing
```

```python
# ◾ Tokenization is used in natural language processing to split 
# paragraphs and sentences into smaller units that can be 
# more easily assigned meaning.

# ◾ The first step of the NLP process is gathering the data (a 
# sentence) and breaking it into understandable parts (words).                                                       
```

```python
# Why tokenization?
#<-------------------->
# > Jb bhi hm kisi data ko kisi algorithm ke andr ya phir neural network ke andr dete hai.
# to us time hmara data numerical format me hota hai.Because The machine can not understand
# numerical data.
# to hm jb machine ko texual data ki jgah numerical data dete hai.
# to numerical data bhi kisi na kisi word se milker bna hota hai.

# > ab agr hm yha pe pure paragraph ko use krke usko numerical word
# ke andr convert krenge that is a very big task.

#  >   Instead of giving whole paragraph to machine we break into the 
# paragraph into small small parts.

# > After breaking we remove the unusefull things like
# percentage, dot, semicolon.
#  and in word ko remove krne ke liye bhi hmare pass jo sentence hai,
# ya paragraph hai unhe break krna pdta hai.
# we can not remove these unusefull things without breaking.


# In short
#<----------->
# Tokenization is used to break down text into smaller units (tokens) 
# like words or phrases, making it easier to analyze,
# process, and understand for tasks like natural language processing or text analysis.


# tokenization is the basic step which we can use to break sentence into words and 
# that we can remove unusual things.
# This is the first step when we use tokenizing.
```

```python
from nltk.corpus import stopwords
from string import punctuation

# Define the stop word removal function
def stop_word_removal(word):
    stop = stopwords.words("english")
    stop_word_list = list(punctuation) + stop
    
    # Split words
    splitting = word.split()
    print(f"splitting_word: {splitting}")
    
    # Remove stop words and punctuation
    removing = [i for i in splitting if i.lower() not in stop_word_list]
    print(f"removing_stop_word: {removing}")
    
    # Join the words back into a sentence
    joining = " ".join(removing)
    print(f"joining: {joining}")
    return joining

# Sample text (without converting to lowercase here)
text = "This IS an Apple"

# Call the function
result = stop_word_removal(text)
print(result)  # Final result
```

## n-grams

▪️ N-grams are continuous sequences of words or symbols or tokens in a document.
In technical terms.
▪️ They can be defined as the neighboring sequences of items in a document.

<h2> N-gram Language Model</h2>

<ul>
    <li>After generalizing, the above equation can be calculated as:
        <ul>
            <li><strong>Word sequences</strong><br>
                \( w_1^n = w_1, \dots, w_n \)
            </li>
            <li><strong>Chain rule of probability</strong><br>
                \[
                P(w_1^n) = P(w_1) P(w_2 | w_1) P(w_3 | w_1^2) \dots P(w_n | w_1^{n-1}) = \prod_{k=1}^n P(w_k | w_1^{k-1})
                \]
            </li>
            <li><strong>Bigram approximation</strong><br>
                \[
                P(w_1^n) \approx \prod_{k=1}^n P(w_k | w_{k-1})
                \]
            </li>
            <li><strong>N-gram approximation</strong><br>
                \[
                P(w_1^n) \approx \prod_{k=1}^n P(w_k | w_{k-N+1}^{k-1})
                \]
            </li>
        </ul>
    </li>
</ul>
