# mycode) Understanding ? , *, +

```python
# ?: Matches zero or one occurrence  
#<---------------------------------------->

import re
a = "ca cha chha coa"
match = re.findall(r"ch?a",a)
match

# ca      -> ca  (0 h ok)
# cha -> cha -> (1 h ok)

# chha    -> no output (2 h not ok)
# coa     -> no output (any character other than h is not ok)
```

```python
# *: Any number of occurrences (including 0 occurrences)
#<---------------------------------------------------------->

# //* matches zero or any occurance
import re
a = "ca cha chha coa"
match = re.findall(r"ch*a",a)
match

# ca      -> ca  (0 h ok)
# cha -> cha -> (1 h ok)
# chha    -> no output (2 h also ok)

# coa     -> no output (any character other than h is not ok)
```

```python
# +: One or more occurrences

import re
a = "ca cha chha coa"
match = re.findall(r"ch+a",a)
match

# ca      -> ca  (charcter before + sign has zero occurance so not match)

# cha -> cha -> (1 occurance ok match)
# chha    -> no output (2 occurance also ok)

# coa     -> no output (any character other than h is not ok)
```

```python
# comparision (?, *, +)
#<------------------------->

# string = cha
# ? -> ch?a (if zero or one occurance of character before sign?  then match/print)
# * -> ch*a (if zero or any occurance of character before sign? print)
# + -> ch+a (if one or any occurance of character before sign+ then it will print)

# here, character before sign is h.

# so note that + plus me agr plus sign se pehle wala character string hona compulsory hai
# wrna wo wala string print nhi hoga. becoz isme sign se pehle wale
# character ka ek occurance hona compulsory hai.
```

# Special Sequences in RegEx

```python
# ● Special sequences do not match for the actual 
# character in the string instead it tells the specific 
# location in the search string where the match must 
# occur.

# ● It makes it easier to write commonly used patterns.
```

```python
# Special                 Description
# Sequence
# ----------------------------------------------------------------------------
# \A              Matches if the string begins with the gives character

# \b              Matches if the word begins or ends with the
#                 given character. \b(string) will check for the 
#                 beginning of the word and (string)\b will 
#                 check for the ending of the word.

# \B              It is the opposite of the \b i.e. the string 
#                 should not start or end with the given regex. 

# \d               Matches any decimal digit, this is equivalent 
#                  to the set class [0-9]

# \D               Matches any non-digit character, this is 
#                  equivalent to the set class [^0-9] 

# \s               Matches any whitespace character. 

# \S               Matches any non-whitespace character 

# \w               Matches any alphanumeric character, this is 
#                  equivalent to the class [a-zA-Z0-9_]

# \W               Matches any non-alphanumeric character.

# \Z               Matches if the string ends with the given regex


# \d ki defination me decimal ka mtlb decial point nhi hai.
# it means decimal digit/place number means 0 to 9
# in simple word \d gives the numbers.
```

# RegEx Sets

```python
# A set is a set of characters inside a pair of square 
# brackets [] with a special meaning.
```

```python
# set           Description
# ---------------------------------------------------------------------------

# [atx] Returns a match where one of the specified characters (a, t, or x) are present.
# [a-h] Returns a match for any lower case character, alphabetically between a and h.
# [^atx] Returns a match for any character EXCEPT a, t, and x.
# [0123] Returns a match where any of the specified digits (0,1,2 or 3) are present.
# [0-9]  Returns a match for any digit between 0 and 9.
# [0-7][0-9] Returns a match for any two-digit numbers form 00 and 79.
# [a-zA-Z] Returns a match for any character alphabetically between a and z, lower case and upper case.
# [+]      In sets, +, *, ., |, (), $, {} has no special meaning, so [+] means: returns a match for any + character in the string.


# [^atx] -> means caret atx                         
```

```python
# re.findall()
# --------------
# Return all non-overlapping matches of pattern in string, as a list of 
# strings. The string is scanned left-to-right, and matches are returned 
# in the order found.

# /hello
# ------>  left to right/


# re.compile()
# --------------
# Regular expressions are compiled into pattern objects, which have
# methods for various operations such as searching for pattern
# matches or performing string substitutions.


# re.split()
# --------------
# Split string by the occurrences of a character or a pattern, upon 
# finding that pattern, the remaining characters from the string are 
# returned as part of the resulting list.


# re.sub():
# --------------
# The method returns a string where matched occurences are 
# replaced with the content of replace variable.


# re.subn():
# --------------
# subn() is similar to sub() in all ways, except in its way of providing 
# output. It returns a tuple with a count of the total of replacement and 
# the new string rather than just the string. 


# re.escape():
# ---------------
# Returns string with all non-alphanumerics backslashed, this is useful 
# if you want to match an arbitrary literal string that may have regular
# expression metacharacters in it.


# re.search()
# ---------------
# This method either returns None (if the pattern doesn't match), or a  
#     re. This method stops after the first match, so this is best suited for 
# testing a regular expression more than extracting data.
```

## //without re.compile() vs re.compile()

```python
# Using re.compile() is more efficient and can lead to faster code execution, 
# especially when the same pattern is reused multiple times, like in a loop.
# see example below 👇:
```

```python
# Ex:1 Without re.compile()
import re

aaa = "Alice scored 95, Bob scored 88, Carol scored 92"
for i in range(3):
    result = re.findall(r"\d+", aaa)  # Finds all numbers in the string
    print(result)
```

```python
# Ex:2 With re.compile()

import re

aaa = "Alice scored 95, Bob scored 88, Carol scored 92"
pattern = re.compile(r"\d+")
for i in range(3):
    result = pattern.findall(aaa)  # Using the compiled pattern
    print(result)
```

# Match Object in RegEx

```python
# A Match object contains all the information about the 
# search and the result and if there is no match found then 
# None will be returned.

# //explanation:
#<---------------->
# match object like when we use search function.
# then search function ke sath hmare pass jitne bhi details aa jate hai.
# like string details, ya hm kon se regular expression use kr rhe hai uske details,
# starting index, ending index, span.

# //In short
#<---------->
# match ka ye kam hai ki ye search ke andr jitne bhi information hai.
# match object use apne pass rkhta hai.
# and then if we want to get different different information.
# then we can get these different different information by applying function.
```

```python
# code 1)
import re

a = "John has scored 98 marks"

match = re.search("\d+",a) # \d+ means reoccuring values
print(match)

# //Now we will use its functions
print(match.re) # means which regular expression are use in this code.
print(match.string) # obtaining string (wo string jiske upar match kr rhe hai)
print(match.start()) # stating span
print(match.end())   # ending span
print(match.span())  # start, end value/index both
print(match.group()) # group means match value



# explanation
#<---------------->
# > \d is a regex pattern that matches any digit (0-9).
# > + is a quantifier that ensures the preceding pattern (\d) is matched one or more times. This means it will match sequences of digits, whether they are single or multi-digit numbers.
# > Therefore, the pattern \d+ will match the first occurrence of digits in the string a. In this case, it will match the number 98.

# In simple + plus ka use krne se wo matching character to ayega.
# but wo character jis word me hai wo pure word ke sath ayega.
```

# Phone Number and Email Verification and Web Scrapping Using Regex

### 1) Phone number verification

```python
# code 1.1)
import re

phn = "222-444-8787" # //this is valid phone no.

if re.search("\d{3}-\d{3}-\d{4}",phn):
    print ("it is a verified number")
else:
    print ("incorrect number")

# I f the phn no. satisfied the pattern condition then it is a verified number.
```

```python
# code 1.2)
import re

phn = "222-4a4-8787"

if re.search("\d{3}-\d{3}-\d{4}",phn):
    print ("it is a verified number")
else:
    print ("incorrect number")
```

### 2) email verification

```python
# Email refers to an electronic mail address that follows a specific format:
#<------------------------------------------------------------------------------>

# username@domain


# > username: The part before the @ symbol. It identifies the user.
# > domain: The part after the @ symbol, like gmail.com, yahoo.com, or example.co.


# Common parts of an email:
# #<-------------------------->
# 1. Service Providers (Domains):
#    .gmail.com: For Gmail (Google).
#    .yahoo.com: For Yahoo Mail.
#    .outlook.com: For Outlook (Microsoft).
#    .example.co: This could be for companies, organizations, or other services.

                             
# 2. Top-Level Domains (TLDs):
# #<----------------------------->
#     .com: Commercial organizations.
#     .org: Non-profit organizations.
#     .net: Network-related services.
#     .co: Common for companies or Colombia (geographical TLD).

        
# Examples:
# . abc@gmail.com: A valid Gmail address.
# . xyz@yahoo.com: A valid Yahoo address.
# . john.doe@company.co: A valid corporate email.

        
# So, when you say email, you're talking about the entire structure (username@domain.com). Domains like gmail.com, yahoo.com, or .co are just parts of different email providers or organizations. All valid combinations following the email format are considered email addresses
```

```python
email = "john78@gmail.com john@.com    david.989@yahoo.com"

print(re.findall("[\w._%]{0,20}@[\w-].[A-Za-z]{2,3}",email))

print(len(re.findall("[\w._%]{0,20}@[\w-].[A-Za-z]{2,3}",email))) # length(len) gives how many matches
```

### 3) Web Scrapping

```python
# Importing 2 required module
#<----------------------------->

import urllib.request
from re import findall

url = "https://www.summet.com/dmsi/html/codesamples/addresses.html"

a = urllib.request.urlopen(url)
html = a.read()
htmlstr = html.decode()
phn = re.findall("\(\d{3}\) \d{3}-\d{4}",htmlstr)

for i in phn:
    print(i)
```

# Email Validation in Python (Using RegEx)

# TO DO understanding

```python
# conditions
#<-------------->
# a-z //(alphabet a to z, all small)
# 0-9 
# ._ time 1
# @ time 1
# . 2,3   // .com, .in, .co


# Notes
#<--------->
# a-z means first character should be alphabet and in lower case in email.

# Note1) dot and underscore -> @ at the rate se pehle only 1 time hi hona chahiye.
# more than 1 hua. to wo wrong email. hota hai.

# Note2) . 2,3  -> dot ki position piche se 2nd ya 3rd pe hona chahiye.
# // .com, .in, .co
```

```python
import re
email_condition="^[a-z]+[\._]?[a-z 0-9]+[@]\w+[\.]\w{2,3}$"
user_email=input(' Enter your Email : ')

if re.search(email_condition,user_email):
    print("Right Email ")
else:
    print("Wrong Email ")


# explanation

# 1) ^ or startswith -> 1st character should be alphabet and in lower case in email.
# 2) ^[a-z] -> means startswith lower case alphabet
# 3) + -> means join concatenate. it is used to merge small small condition so we can make complete condition.
# 4) \. -> dot is escape sequence character or special character. so here backslash is used to drop special character.
# it is treated as dot not special charater if we used backslash.


# 5) ? means -> dot 0 ya 1 hi hona chahiye. 0 means kuch bhi ni to bhi shi hai. but more than 1 ni hone chahiye.
# other wise string will not match with our pattern.

    
# 6) [a-z 0-9] -> also [a-z]+[0-9] hm group of words and no. ko ek sath bhi de skte hai and alg alg bhi
```
